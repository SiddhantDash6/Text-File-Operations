```python
import csv  

number_lines = sum(1 for row in (open(r"C:\Users\hp\Desktop\Task2\katahane\Katahane01.csv", newline='\n'))) 
for row in range(1,number_lines):
    if row % 2 == 0:
        listOfJapaneseLine.append(row)
    else:
        listOfEnglishLine.append(row)
```

```python
import csv  

outfile = open("Katahane01write.csv",'w')
with open(".\Katahane01.csv",'r', newline = '') as f:
    reader = csv.reader(f, delimiter = " ")
    for row in reader:
        if row[0].startswith("//"):
            print(listOfEnglishLine.append(row))
        else:
            print(listOfJapaneseLine.append(row))
    line = "{},{}\n".format(listOfEnglishLine, listOfJapaneseLine)
    outfile.write(line)
outfile.close()
```
