```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://web.archive.org/web/20180905100235/https://tlwiki.org/index.php?title=Katahane"
```

```python
html_content = requests.get(url).text
```

```python
soup = BeautifulSoup(html_content, "lxml")
print(soup.prettify())
```

```python
for link in soup.find_all("a"):
    print("Inner Text: {}".format(link.text))
    print("Title: {}".format(link.get("title")))
    print("href: {}".format(link.get("href")))
```

```python
Katahane_table = soup.find("table", attrs={"border":"1"})
Katahane_table_data = Katahane_table.tbody.find_all("tr", attrs={"valign":"top"}) 

data = []
for td in Katahane_table_data[1].find_all("td"):
    data.append(td.b.text.replace('\n','').strip())

print(data)
```
